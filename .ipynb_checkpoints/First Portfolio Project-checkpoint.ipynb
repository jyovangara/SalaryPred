```python
#The below function should be very generic so that it is usable for a new Dataset

def plot_feature(df, col):
    #make plot for each feature
    #Left - the distribution of Samples on the feature
    #Right - the dependency f Salary on the feature
    plt.figure(figsize = (14, 6))
    plt.subplot(1,2,1)
    if df[col].dtype == 'int64':
        df[col].value_counts().sort_index().plot()
    else:
        #change the caregorical variable to category type and order their level by mean salary
        #in each category
        mean = df.groupby(col)['salary'].mean()
        df[col] = df[col].astype('category')
        levels = mean.sort_values().index.tolist()
        df[col].cat.reorder_categories(levels, inplace = True)
        df[col].value_counts().plot()
    plt.xticks(rotation = 45)
    plt.xlabel(col)
    plt.ylabel('Counts')
    plt.subplot(1,2,2)
    
    if df[col].dtype == 'int64' or col == 'companyId':
        #plot the mean salary for each category and fill between the (mean - std and mean + std)
        mean = df.groupby(col)['salary'].mean()
        std = df.groupby(col)['salary'].std()
        mean.plot()
        plt.fill_between(range(len(std.index)), mean.values-std.values, mean.values + std.values, \
                         alpha = 0.1)
    else:
        sns.boxplot(x = col, y = 'salary', data=df)
    plt.xticks(rotation = 45)
    plt.ylabel('Salaries')
    plt.show()
    
```

```python
plot_feature(train_df, 'companyId')
```

```python
plot_feature(train_df, 'jobType')
```

```python
plot_feature(train_df, 'degree')
```

```python
plot_feature(train_df, 'major')
```

```python
plot_feature(train_df, 'industry')
```

```python
plot_feature(train_df, 'yearsExperience')
```

```python
plot_feature(train_df, 'milesFromMetropolis')
```

```python
def encode_label(df, col):
    #encode the categories  using avg salary for each category to replace label
    cat_dict ={}
    cats = df[col].cat.categories.tolist()
    for cat in cats:
        cat_dict[cat] = train_df[train_df[col] == cat]['salary'].mean()
    df[col] = df[col].map(cat_dict)
```

```python
for col in train_df.columns:
    if train_df[col].dtype.name == "category":
        encode_label(train_df, col)
```

```python
#Correlation between selected feature and response
#JObid is discarded as it is unique
fig = plt.figure(figsize = (12, 10))
features = ['jobId', 'companyId', 'jobType', 'degree', 'major', 'industry', 'yearsExperience', 'milesFromMetropolis']
sns.heatmap(train_df[features + ['salary']].corr(), cmap='Blues', annot = True)
plt.xticks(rotation=45)
plt.show()
```
